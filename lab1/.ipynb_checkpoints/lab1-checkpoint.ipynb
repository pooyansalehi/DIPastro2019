## Exercise 1 - Read Images from scikit-image

Import necessary packages into python

```python
import matplotlib.pyplot as plt
import matplotlib
from skimage import data
from skimage import io
```

Load in image from database for Lab sessions:

```python
image = data.astronaut()
```

Display images to python

```python
plt.imshow(image)
```

Extract information from image

```python
image.max()
image.min()
image.mean()
image.ndim
image.size
image.view()
image.shape
image.dtype
```

Convert the image of the astronaut to a grayscale image

```python
from skimage.color import rgb2gray
```

```python
original = data.astronaut()
```

```python
grayscale = rgb2gray(original)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

Create the inverse of the grayscale image of the astronaut

```python
import numpy as np
from skimage import util 
```

```python
igr = util.invert(grayscale)
```

```python
plt.imshow(igr, cmap='gray')
```
